```
data = pd.read_csv("output.csv")
data.head()
```

```
data_1 = data.loc[:,['season_year','team_name','Opp','Score_Tm','Score_Opp','Off_1stD','Def_1stD','Off_TotYd','Def_TotYd','Off_PassY','Def_PassY','Off_RushY','Def_RushY','Off_TO','Def_TO','EP_Offense','EP_Defense','ER_Sp.Tms']]
data_1
```

```
data_S = data_1.loc[data_1['team_name'] =='San Francisco 49ers']
print(data_S)
data_K = data_1.loc[data_1['team_name'] =='Kansas City Chiefs']
print(data_K)
```

```
data_K_2019 = data_K.loc[data_K['season_year'] == 2019]
pred_Score_Tm = data_K_2019['Score_Tm'].mean()
pred_Score_Opp = data_K_2019['Score_Opp'].mean()
pred_Off_1stD = data_K_2019['Off_1stD'].mean()
pred_Def_1stD	= data_K_2019['Def_1stD'].mean()
pred_Off_TotYd = data_K_2019['Off_TotYd'].mean()
pred_Def_TotYd = data_K_2019['Def_TotYd'].mean()
pred_Off_PassY = data_K_2019['Off_PassY'].mean()
pred_Def_PassY = data_K_2019['Def_PassY'].mean()
pred_Off_RushY = data_K_2019['Off_RushY'].mean()
pred_Def_RushY = data_K_2019['Def_RushY'].mean()
pred_Off_TO = data_K_2019['Off_TO'].mean()
pred_Def_TO = data_K_2019['Def_TO'].mean()
pred_EP_Offense	= data_K_2019['EP_Offense'].mean()
pred_EP_Defense	= data_K_2019['EP_Defense'].mean()
pred_ER_Sp_Tms = data_K_2019['ER_Sp.Tms'].mean()
```

```
data_3 = data_K.append({'season_year':2019, 'team_name':'Kansas City Chiefs', 'Opp':'San Francisco 49ers', 
                        'Score_Tm':pred_Score_Tm, 'Score_Opp':pred_Score_Opp, 'Off_1stD':pred_Off_1stD,	
                        'Def_1stD':pred_Def_1stD,	'Off_TotYd':pred_Off_TotYd,	'Def_TotYd':pred_Def_TotYd,
                        'Off_PassY':pred_Off_PassY,	'Def_PassY':pred_Def_PassY,	'Off_RushY':pred_Off_RushY,
                        'Def_RushY':pred_Def_RushY,	'Off_TO':pred_Off_TO,	'Def_TO':pred_Def_TO,	
                        'EP_Offense':pred_EP_Offense,	'EP_Defense':pred_EP_Defense,	'ER_Sp.Tms':pred_ER_Sp_Tms}, ignore_index=True)
data_3

########################
#feature improve 1


```

```
##############X = data_K.iloc[:,4:]
X = data_K[['Off_TotYd','Off_RushY','EP_Defense','Off_PassY','Def_RushY','Def_TotYd','EP_Offense','Def_PassY']]
y = data_K['Score_Tm'].values
print(X,y)
#type(y)
#type(X)
```

```
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.model_selection import cross_val_score
from sklearn.utils import resample

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)
#print(X_test)
#type(X_test)
```

```
import xgboost as xgb
from xgboost import plot_importance
from matplotlib import pyplot as plt

regr = xgb.XGBRegressor(
                 colsample_bytree=0.2,
                 gamma=0.0,
                 learning_rate=0.05,
                 max_depth=6,
                 min_child_weight=1.5,
                 n_estimators=7200,                                                                  
                 reg_alpha=0.9,
                 reg_lambda=0.6,
                 subsample=0.2,
                 seed=42,
                 silent=1)

regr.fit(X_train,y_train)

# Run prediction on training set to get a rough idea of how well it does.
y_pred = regr.predict(X_test)
print(y_pred)
print('RMSD:',mean_squared_error(y_test,y_pred)**0.5) #均方根误差，RMSD,root mean square deviation，是观测值与真值偏差的平方和与观测次数m比值的平方根，用来衡量观测值同真值之间的偏差
# 显示重要特征
plot_importance(regr)
plt.show()
print(X_test)
```
